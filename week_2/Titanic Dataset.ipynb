```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_titanic = pd.read_csv('./Titanic-Dataset.csv')
df_titanic.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df_titanic, x='Age', bins=40, kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print("Distribution of Embarked Column:")
plt.figure(figsize=(8, 6))
sns.countplot(data=df_titanic, x="Embarked")
plt.title("Embarked Distribution of Titanic Passenger")
plt.xlabel("Embarked")
plt.ylabel("Frquencey")
plt.show()
```

```python
age_median = df_titanic['Age'].median()
df_titanic['Age'].fillna(age_median, inplace=True)
```

```python
embarked_mode = df_titanic['Embarked'].mode()[0]
df_titanic['Embarked'].fillna(embarked_mode, inplace=True)
```

```python
df_titanic.drop(columns=["Cabin"], inplace=True)
```
